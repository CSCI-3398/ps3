### Pitch extraction in parselmouth

This is a quick little Jupyter notebook demostrating how to extract pitch using parselmouth. You will have to install parselmouth, which you can do within Anaconda through its package manager (if you're using Anaconda) or from a terminal with `pip3 install praat-parselmouth` (if you are using your own installation of Python).

For more example scripts, check out [the parselmouth docs](https://parselmouth.readthedocs.io/en/stable/examples.html) and [this helpful repo of scripts](https://github.com/drfeinberg/PraatScripts)

```python
# Import statements

import glob
import parselmouth
from parselmouth.praat import call
```

```python
# Go through all the wav files in the data folder and print out the mean pitch

for wav_file in glob.glob("/Users/emilypx/Desktop/ps3/data/*.wav"):
    sound = parselmouth.Sound(wav_file)
    pitch = call(sound, "To Pitch", 0, 75, 600) 
    meanF0 = call(pitch, "Get mean", 0, 0, "Hertz")
    print(wav_file, meanF0)


```
